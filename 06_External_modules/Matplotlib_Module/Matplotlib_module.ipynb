<h1>Matplotlib Module</h1>

<h3>Example 1: Draw a line from position (1,4) to (2,5) then to (3,1)</h3>

plot() function : to draw a line from point to point

Syntax - plot(parameter1, paramter2)

parameter1 - an array containing x-axis points

parameter2 - an array containing y-axis points

```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[4,5,1])
plt.show()
```

<h3>Example 2: Draw a line from from (1,1)  to (8,3)</h3>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8])
y = np.array([1,3])

plt.plot(x,y)
plt.show()
```

<h3>Example 3: Using Marker, linestype, color, linewidth</h3>

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[4,5,1], marker = 'o', ls = '--', color = 'red', lw = 4)
plt.show()
```

<h3>Example 4: Draw two lines</h3>

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1,3,5,7])
y1 = np.array([2,5,3,8])

y2 = np.array([5,1,8,3])

plt.plot(x1,y1, color = 'green')
plt.plot(y2)
plt.show() 
```
